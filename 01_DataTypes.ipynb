## Data type <a class="anchor" id="first-bullet"></a>

There are several data types in Python. The following are the most relevant ones


- Integer (int)
- Float (float)
- Boolean (bool)
- String (str)

To verify the type of data, the type() function can be used

#### Integers

```python
numero = 232
type(numero)
```

#### Floats

```python
numero_2 = 232.232
type(numero_2)
```

#### Booleans

```python
bool_1= True
bool_2 = False
type(bool_1)
```

#### Strings

```python
cadena = "curso"
type(cadena)
```

## Data type conversion <a class="anchor" id="second-bullet"></a>

```python
# float --> int

a = int(3.45)
a
```

```python
# string --> int
b = int("34")
b
```

```python
# int --> float
c = float(a)
```

```python
# string --> float

d = float("3.5")
d
```

```python
# int --> str
e = str(345)
e
```

```python
# float --> str
f = str(3.45)
f
```

## Input function <a class="anchor" id="third-bullet"></a>

This is an interactive way to input data. 

To avoid errors, it is recommended to specify de type of data being required. For example str(input()) is specifying that the expected data is a string. If the user enters a different type of data, the code will immediately let them know. 

```python
#String
str_1 = str(input("Type your username"))
print("Welcome", str_1)
```

```python
#Integer

int_1 = int(input("Type your favorite integer number"))
a =  2*int_1

print("Two times your favorite integer number is",a)

```

```python
#Float
int_1 = float(input("Type your favorite float number"))
a =  2*int_1

print("Two times your favorite float number is",a)
```
