#### Concatenate <a class="anchor" id="first-bullet"></a>

```python
## Concatenate two strings

name = 'Bob'
lastname = 'Brown'
fullname = name + lastname

#no space
print('Full name_1 = ', fullname)

#space 
print('Full name', name,lastname)
```

```python
#Concatenate string and number
age = 50

print('Full name', name, lastname, f"age: {age}")

'''f concatenates strings and other type of data. Notice that both data are
inside the quotation marks '''
```

```python
message = 'your age is'
print(f"{message} {age}")
```

#### Upper <a class="anchor" id="second-bullet"></a>

Transforms all the letters into capital letters

```python
name_1 = 'Bob'
name_1 = name_1.upper()
print(name_1)
```

#### Lower <a class="anchor" id="third-bullet"></a>

Transforms all the letters into lower case

```python
name_2 = name_1.lower()
print(name_2)
```

#### Capitalize <a class="anchor" id="fourth-bullet"></a>

Transforms the first letter of the string into a capital letter

```python
name_3 = name_2.capitalize()
print(name_3)
```

#### Replace <a class="anchor" id="fifth-bullet"></a>

Replace the letter of a string with the new specified letters

```python
str_1 = "Data analysis"
str_1 = str_1.replace("Data", "Introduction to data")
str_1
```

#### Title <a class="anchor" id="sixth-bullet"></a>

Capitalize the first letter after a space

```python
str_2 = str_1.title()
str_2
```

#### Split <a class="anchor" id="seventh-bullet"></a>

Splits the string after the specified symbol into a list of strings

```python
str_3 = str_2.split('o')
print(str_3)
print(type(str_3))
```
