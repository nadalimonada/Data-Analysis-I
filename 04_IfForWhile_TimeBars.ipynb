## Conditionals <a class="anchor" id="first-bullet"></a>

```python
num = input("Enter an integer")

if type(num) == int:

    if num > 0:
        print("The number is positive")
    
    elif num < 0:
        print("The number isn't positive")
        
    else:
        print("The number is neutral")

else: 
    print("The number was not an integer")
```

## Time library <a class="anchor" id="second-bullet"></a>

This library notifies how long it took the cell to run 

To do this, '%%time' should be written at the beginning of the cell

```python
import time
```

```python
%%time
age = float(int(input("Enter your age: ")))

if age >= 18:
    print("Adult")
if age < 18:
    print("Under-aged")
```

## For loops <a class="anchor" id="third-bullet"></a>

```python
list_1 = [2,4,6,8]

for element in list_1:
    print(element**2)
```

#### Break

Ends the loop completely depending on the conditioal

```python
for element in range(0, 100):
    print(element)

    if element == 6:
        break
```

#### Continue

The continue keyword is used to end the current iteration in a for loop (or a while loop), and continues to the next iteration.

```python
# Skip the iteration if the variable i is 3, but continue with the next iteration
# (Doesn't print 3):
for i in range(9):
  if i == 3:
    continue
  print(i)
```

#### Pass

The pass statement is used as a placeholder or filler for future code.

When the pass statement is executed, nothing happens, but you avoid getting an error when empty code is not allowed.

Empty code is not allowed in loops, function definitions, class definitions, or in if statements.

```python
for elemento in range(0, 5):
    if elemento == 2:
        pass
        
    print(elemento)
```

## tqdm libary <a class="anchor" id="fourth-bullet"></a>

Tqdm is a Python library used for creating progress bars. It can be used to visualize the progress of loops and other iterable processes. The library's name is derived from the Arabic word "taqaddum," which means "progress.

```python
from tqdm import tqdm
```

```python
for elemento_1 in tqdm(range(0, 1000)):
    for elemento_2 in range(0, 100):
        pass
        
```

## While loop <a class="anchor" id="fifth-bullet"></a>

With the while loop we can execute a set of statements as long as a condition is true

```python
i = 1
while i < 6:
  print(i)
  i += 1
```
