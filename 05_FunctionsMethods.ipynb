## Functions <a class="anchor" id="first-bullet"></a>

A function is a block of code which only runs when it is called.

A function can return data as a result or just do an action like plotting 

A function helps avoiding code repetition.


#### No data return

```python
def sum1(num_1, num_2):

    Sum= num_1 + num_2
    print(Sum)
```

```python
sum1(3, 2)
```

#### Single data return

```python
def sum2(num_1, num_2):

    Sum= num_1 + num_2
    return Sum
```

```python
total_sum = sum2(5,8)
print(total_sum)
```

#### Multiple data return

```python
def add_subs_multp(num_1, num_2):

    add = num_1 + num_2
    subs = num_1 - num_2
    multp = num_1*num_2

    return add, subs, multp
```

```python
# List with all the returned data
results = add_subs_multp(8,5)

print('Addition:', results[0])
print('Substraction',results[1])
print('Multiplication',results[2])
```

```python
# Individual returned data 

a, s, m = add_subs_multp(10,5)

print('Addition:', a)
print('Substraction', s)
print('Multiplication', m)
```
