## Lists <a class="anchor" id="first-bullet"></a>

````python
list_1.extend([2,1])
print("Extend method",list_1)

'''Please notice that if you try to to the same with the append method,
a sub-list would be added'''

list_2 = list_1.copy()
list_2.append([3,4])
print("Append method",list_2)
````

## Tuple <a class="anchor" id="second-bullet"></a>

It works just like a list with the exemption that tuples can't be modified. Let's consider the following situation

``` Python
list_1 = [a,b,c]

list_2 = list_1

list_1[0] = d

print(list_1, list_2)

[d,b,c] , [d,b,c] 
``` 

In this case, as I modified list_1, there will be a inter-process memory sharing that will modify list_2.

In tuples, every tuple is independen and the modification of the original one won't modify the copy. 

````python
tuple_1 = ('a', 'b', 'c')
tuple_2 = tuple_1

print('tuple 1:', tuple_1)

print('tuple 2:', tuple_2)
````

````python
#it is necessary to convert the tuple into a list due to the immutable nature 
# of the tuple
 
temp_list = list(tuple_1)
temp_list[0] = 'd'
tuple_1 = tuple(temp_list)

print('tuple 1:', tuple_1)

print('tuple 2:', tuple_2)
````

## Dictionary <a class="anchor" id="third-bullet"></a>

Just like lists, dictionarys can storage different data types. The difference is that dictionaries have the strcutre `key: value`. 

``` Python
dict_values = {
    "key": value,
    "key_2": value,
       .       .
       .       .
       .       .
}

```

````python
#The next example is a nested dictionary, {key: [{Key:[value], Key:[value]}]
dic = {
    "States": [{
    "California": {
        "Cities": ["Calabasas", 'Campbell', 'Carpintería'],
        "Capital": ["Sacramento"]
    }, 
        "Texas": {
            "Cities": ['Houston', 'Dallas', 'San Antonio'],
            "Capital": ['Austin']
        }
    }]
}

dic
````

#### Keys
